### Experimenting with time-based stuff w/ and w/o parallelism

```python
%%time

from dask import delayed, compute

def my_function(x):
    import time
    time.sleep(2)  # Simulate a time-consuming task
    return x ** 2

tasks = [delayed(my_function)(i) for i in range(10)]
results = compute(*tasks)

print(results)
```

```python
def cube(x):
    import time
    hoststring = socket.gethostname()
    time.sleep(2)  # Simulate a time-consuming task
    return (x ** 3, hoststring)
    
workers = client.scheduler_info()['workers']
worker_addresses = list(workers.keys())
target_worker = worker_addresses[0]

# Submit a task to the specific worker
future = client.submit(cube, 8, workers=[target_worker])
result = future.result()
print(result)
```
